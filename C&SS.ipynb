```python
#@title Ceasar Cipher (Encrypt)

#encrypt
def encrypt_text(plaintext,n):
    ans = ""
    for i in plaintext:
        if i == " ":
            ans+=" "
        #Uppercase
        elif (i.isupper()):
            ans += chr((ord(i) + n-65) % 26 + 65)
        #lowercase
        else:
            ans += chr((ord(i) + n-97) % 26 + 97)
    
    return ans

plaintext = "HELLO EVERYONE"
n = 5
print("Plain Text : " + plaintext)
print("Shift pattern : " + str(n))
print("Cipher Text : " + encrypt_text(plaintext,n))
```

```python
#@title Ceasar Cipher (Decrypt)
def decrypt():
    
    encrypted_message = input("Enter the message i.e to be decrypted: ").strip()
    
    letters="abcdefghijklmnopqrstuvwxyz"
    
    #enter the key value to decrypt
    k = int(input("Enter the key to decrypt: "))
    decrypted_message = ""

    for ch in encrypted_message:

        if ch in letters:
            position = letters.find(ch)
            new_pos = (position - k) % 26
            new_char = letters[new_pos]
            decrypted_message += new_char
        else:
            decrypted_message += ch
    print("Your decrypted message is:\n")
    print(decrypted_message)

decrypt()
```

```python
#@title Playfair Cipher
import numpy as np

#convert to lower case
def lower_Case(text):
    return text.lower()

#remove space
def remove_Space(text):
    ts = []
    for i in text:

        # if space then goto next
        if i == " ":
            continue

        # else continue
        else:
            ts.append(i)
    return ts

#generate matrix
def mymatrix(key,alphList):
    mat = []
    for i in key:
        if i not in mat:
            mat.append(i)
    for i in alphList:
        if i not in mat:
            mat.append(i)
    mat = np.reshape((mat),(5,5))
    return mat

#replace duplicate with x    
def FillerLetter(text):
	k = len(text)
    
    #if len(text) == even
	if k % 2 == 0:
		for i in range(0, k, 2):
            #if consecutive letters same replace sec with x 
			if text[i] == text[i+1]:
				new_word = text[0:i+1] + [str('x')] + text[i+1:]
				new_word = FillerLetter(new_word)
				break
			else:
				new_word = text
    # if len(text)== odd
	else:
		for i in range(0, k-1, 2):
			if text[i] == text[i+1]:
				new_word = text[0:i+1] + [str('x')] + text[i+1:]
				new_word = FillerLetter(new_word)
				break
			else:
				new_word = text
	return new_word


#group two elements
def grp(text):
	grpList = []
	group = 0
	for i in range(2, len(text), 2):
		grpList.append(text[group:i])

		group = i
	grpList.append(text[group:])
	return grpList

#search
def search(mat, x):
	for i in range(5):
		for j in range(5):
			if(mat[i][j] == x):
				return i, j

#row rule
def encrypt_RowRule(mat, i1, j1, i2, j2):
	char1 = ''
	if j1 == 4:
		char1 = mat[i1][0]
	else:
		char1 = mat[i1][j1+1]

	char2 = ''
	if i2 == 4:
		char2 = mat[i2][0]
	else:
		char2 = mat[i2][j2+1]

	return char1, char2

#Col rule
def encrypt_ColumnRule(mat, i1, j1, i2, j2):
	char1 = ''
	if i1 == 4:
		char1 = mat[0][j1]
	else:
		char1 = mat[i1+1][j1]

	char2 = ''
	if i2 == 4:
		char2 = mat[0][j2]
	else:
		char2 = mat[i2+1][j2]

	return char1, char2

#Box rule
def encrypt_BoxRule(mat, i1, j1, i2, j2):
    char1 = ''
    char1 = mat[i1][j2]

    char2 = ''
    char2 = mat[i2][j1]

    return char1, char2

def encrypt(Mat, grpList):
	cipherT = []
	for i in range(0, len(grpList)):
		c1 = 0
		c2 = 0
		i1, j1 = search(Mat, grpList[i][0])
		i2, j2 = search(Mat, grpList[i][1])

		if i1 == i2:
			c1, c2 = encrypt_RowRule(Mat, i1, j1, i2, j2)
			# Get 2 letter cipherText
		elif j1 == j2:
			c1, c2 = encrypt_ColumnRule(Mat, i1, j1, i2, j2)
		else:
			c1, c2 = encrypt_BoxRule(
				Mat, i1, j1, i2, j2)

		cipher = c1 + c2
		cipherT.append(cipher)
	return cipherT

#variables
alphList = ['a','b','c','d','e','f','g','h','i','k','l',
            'm','n','o','p','q','r','s','t','u','v','w','x','y','z']
key = "crazy"
text1 = "Hello W world"

#remove space
rskey = remove_Space(lower_Case(key))
rstext1 = remove_Space(lower_Case(text1))

#generate matrix
keyMatrix = mymatrix(rskey,alphList)

#create grp
grpList = grp(FillerLetter(rstext1))

if len(grpList[-1]) != 2:
	grpList[-1] = grpList[-1]+['z'] #add z to complete pair

#encrypt
CipherList = encrypt(keyMatrix, grpList)
CipherText = ""
for i in CipherList:
	CipherText += i

print("PlainText: ", text1)
print("CipherText: ", CipherText)
```

```python
#@title Implement the Fiestal cipher algorithm. Test your implementation by encrypting end decrypting sample message.

def feistel_cipher(text, key, rounds):

    block_size = len(text) // 2
    left = text[:block_size]
    right = text[block_size:]

    for i in range(rounds):
        new_right = xor(left, feistel_function(right, key))
        left = right
        right = new_right

    # swap left and right and return as ciphertext/plaintext
    return right + left

def feistel_function(text, key):
    return xor(text, key)

def xor(a, b):
    return ''.join(chr(ord(x) ^ ord(y)) for x, y in zip(a, b))

plaintext = "Hello, world!"
key = "secretkey"
rounds = 4

ciphertext = feistel_cipher(plaintext, key, rounds)
decrypted_plaintext = feistel_cipher(ciphertext, key, rounds)

print(f"Plaintext: {plaintext}")
print(f"Ciphertext: {ciphertext}")
print(f"Decrypted plaintext: {decrypted_plaintext}")
```

```python
#@title RC4
def rc4(key, plaintext):
    S = list(range(256))
    j = 0
    out = []

    # Key scheduling algorithm
    for i in range(256):
        j = (j + S[i] + key[i % len(key)]) % 256
        S[i], S[j] = S[j], S[i]

    # Pseudo-random generation algorithm
    i = j = 0
    for char in plaintext:
        i = (i + 1) % 256
        j = (j + S[i]) % 256
        S[i], S[j] = S[j], S[i]
        out.append(chr(ord(char) ^ S[(S[i] + S[j]) % 256]))

    return ''.join(out)

key = b'secretkey'
plaintext = 'Hello, world!'
print(f"Plain Text:{plaintext}")
ciphertext = rc4(key, plaintext)
print(f"Cipher Text:{ciphertext}")

plaintext = rc4(key, ciphertext)
print(f"Decrypted Text:{plaintext}")
```
